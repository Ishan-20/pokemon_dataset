```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('/kaggle/input/pokemon-dataset/Pokemon.csv')
print(df.columns)
df.head(n=10) 
# print(df.columns) 
# df.tail(10)  Returns the last n rows of the DataFrame.
# df.info()   Provides a concise summary of the DataFrame.
# df.shape    Returns the dimensions of the DataFrame (number of rows and columns).
# df.columns  Returns the column labels of the DataFrame.
# df.dtypes    Returns the data types of each column.
# df.isnull().sum()  Checks for missing values and sums them for each column
# df.sort_values(by='HP') Sorts the DataFrame by the specified column.
# df.drop(columns=['Generation'])   Removes specified columns from the DataFrame.
# df.rename(columns={'#': 'NO.'})    Renames columns in the DataFrame.
```

```python
df.columns = df.columns.str.upper()
df['#'] = range(1, len(df) + 1)        # Add a unique key to the '#' column
print(df.columns)
df.head(10)
# df.columns.str.contains('pokemon')
# df.columns.str.len()

```

```python
df[df['LEGENDARY'] == True].head(5)   #Showing the legendary pokemons
```

```python
df = df.set_index('NAME')
df.head(n=10)
```

```python
def clean_index(NAME):
    if 'Mega' in NAME:
        return NAME[NAME.find('Mega'):]
    return NAME
df.index = [clean_index(NAME) for NAME in df.index]
df.head(n=10) 
```

```python
df = df.drop(['#'],axis=1)
df.head(n=10)
```

```python
print('The columns of the dataset are: ',df.columns)
print('The shape of the dataframe is: ',df.shape)
```

```python
df['TYPE 2'] = df['TYPE 2'].fillna(df['TYPE 1'])  #some values in TYPE2 are empty and thus they have to be filled or deleted
has_nulls = df['TYPE 2'].isnull().any()
print(f"\nNumber of NaN values left in 'TYPE 2': {has_nulls}")
df.head(n=10)
```

```python
max_hp_index = df['HP'].idxmax()
max_hp_value = df['HP'].max()
max_hp_name = max_hp_index
print(f"Name: {max_hp_name}, Max HP: {max_hp_value}")

max_defense_index = df['DEFENSE'].idxmax()
max_defense_value = df['DEFENSE'].max()
max_defense_name = max_defense_index
print(f"Name: {max_defense_name}, Max DEFENSE: {max_defense_value}")


max_attack_index = df['ATTACK'].idxmax()
max_attack_value = df['ATTACK'].max()
max_attack_name = max_attack_index
print(f"Name: {max_attack_name}, Max ATTACK: {max_attack_value}")
```

```python
bins=range(0,200,20) #they act as containers
plt.hist(df["ATTACK"],bins,histtype="bar",rwidth=1.2,color='#0ff0ff') #hist() is used to plot a histogram
plt.xlabel('Attack') #set the xlabel name
plt.ylabel('Count') #set the ylabel name
plt.plot()
plt.axvline(df['ATTACK'].mean(),linestyle='dashed',color='red')
plt.show()
```

```python
fire=df[(df['TYPE 1']== 'Fire') | ((df['TYPE 2'])=="Fire")]
water=df[(df['TYPE 1']== 'Water') | ((df['TYPE 2'])=="Water")]
plt.scatter(fire.ATTACK.head(50),fire.DEFENSE.head(50),color='r',label='Fire',marker="*",s=50) #scatter plot
plt.scatter(water.ATTACK.head(50),water.DEFENSE.head(50),color='b',label='Water',s=25)
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.legend()
plt.plot()
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.show()
```
